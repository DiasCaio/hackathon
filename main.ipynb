### Vamos agora começar os preparos para o treino da base

```python
import numpy as np, pandas as pd, math, gc, time, warnings
from sklearn.model_selection import TimeSeriesSplit
warnings.filterwarnings("ignore")


GAP = 1                 # gap entre treino e teste (em semanas) para tentar otimizar o treinamento
TRAIN_FRAC = 0.80       # porcentagem de dados de treino
ROUNDING = 'ceil'       # arredondamento do corte de semanas para cima, de forma que, caso necessário, a gente arrendonde para cima
RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)

# Utilizei anteriormente a otimização de hiper parâmetros e obtive os resultados abaixo. Para que rode mais rápido, vou deixar apenas as
#que desempenharem melhor
CONFIGS = [
    {'lookback': 8, 'units': 64, 'dropout': 0.2, 'loss': 'huber',
     'lr': 1e-3, 'batch_size': 1024, 'epochs': 12},
]
```

### Vamos criar um novo dataframe limpo e garantir que ele está perfeito para usarmos

```python
cols_req = ['pdv','produto','ano_iso','semana','quantity']

# Cópia e coerção de tipos
df_nn = df_modelo_inicial[cols_req].copy()
df_nn['pdv']      = df_nn['pdv'].astype(str)
df_nn['produto']  = df_nn['produto'].astype(str)
df_nn['ano_iso']  = df_nn['ano_iso'].astype(int)
df_nn['semana']   = df_nn['semana'].astype(int)
df_nn['quantity'] = pd.to_numeric(df_nn['quantity'])
df_nn = df_nn.dropna(subset=['quantity'])

# Colapsa duplicatas por (pdv, produto, ano_iso, semana) somando a quantidade
if df_nn.duplicated(subset=['pdv','produto','ano_iso','semana']).any():
    df_nn = (df_nn
             .groupby(['pdv','produto','ano_iso','semana'], as_index=False)['quantity']
             .sum())

# Data âncora ISO: segunda-feira da semana (útil para ordenação temporal correta)
df_nn['week_start'] = pd.to_datetime(
    df_nn['ano_iso'].astype(str) + df_nn['semana'].astype(str).str.zfill(2) + '1',
    format='%G%V%u', errors='coerce'
)
df_nn = df_nn.dropna(subset=['week_start'])

# Ordenação temporal final
df_nn = df_nn.sort_values(['pdv','produto','week_start']).reset_index(drop=True)

print(f"[STAGE 1] df_nn pronto ({len(df_nn):,} linhas)")
```

### Dividindo o dataset em treino e teste

```python
# Semanas únicas ordenadas por tempo
weeks = (df_nn[['ano_iso','semana','week_start']]
         .drop_duplicates()
         .sort_values('week_start')
         .reset_index(drop=True))

W = len(weeks)
# cálculo do tamanho do bloco de treino 
train_target = math.ceil (TRAIN_FRAC * W)

test_size = max(1, W - GAP - train_target)

# TimeSeriesSplit para obter o último corte 80/20 com gap
tss = TimeSeriesSplit(n_splits=3, test_size=test_size, gap=GAP)
for tr_idx, te_idx in tss.split(weeks):
    last_tr_idx, last_te_idx = tr_idx, te_idx

weeks_train = weeks.iloc[last_tr_idx]
weeks_test  = weeks.iloc[last_te_idx]

# Marcação no df
df_nn['split'] = 'train'
mi_test = pd.MultiIndex.from_frame(weeks_test[['ano_iso','semana']])
is_test = pd.MultiIndex.from_frame(df_nn[['ano_iso','semana']]).isin(mi_test)
df_nn.loc[is_test, 'split'] = 'test'

print(f"[STAGE 2] Semanas: {W} | Treino: {len(weeks_train)} | Gap: {GAP} | Teste: {len(weeks_test)}")
```

### Normalizando os dados para deixa-los na mesma "ordem de grandeza", melhorando o desempenho do modelo

```python
# min/max por (pdv, produto) calculados só no treino, para não "vermos o futuro" com os dados de teste
train_minmax = (
    df_nn[df_nn['split']=='train']
      .groupby(['pdv','produto'])['quantity']
      .agg(['min','max'])
      .rename(columns={'min':'ymin','max':'ymax'})
      .reset_index()
)
train_minmax['range'] = (train_minmax['ymax'] - train_minmax['ymin'])

# anexa aos dados
df_nn = df_nn.merge(train_minmax, on=['pdv','produto'], how='left')

# fallback para séries que só aparecem no teste
global_min = df_nn.loc[df_nn['split']=='train','quantity'].min()
global_max = df_nn.loc[df_nn['split']=='train','quantity'].max()
global_range = max(global_max - global_min, 1.0) #garantindo que não vamos dividir por 0

mask_missing = df_nn['ymin'].isna()
df_nn.loc[mask_missing, 'ymin']  = global_min
df_nn.loc[mask_missing, 'ymax']  = global_max
df_nn.loc[mask_missing, 'range'] = global_range

# escala por série
df_nn['quantity_scaled'] = (df_nn['quantity'] - df_nn['ymin']) / df_nn['range']

print(f"[STAGE 3] Escala por série aplicada | y_train min/max globais: {global_min} / {global_max}")
```

### Algumas funções auxiliares para nosso treinamento

```python

# Cria as janelas de treino do nosso conjunto e guarda a escala utilizada para normalizar os dados, de forma que a gente consiga desconverter
# os dados para a versão original no futuro
def build_windows_intrasplit_with_meta(df_src, lookback=8):
    """Treino/val intra-split: gera janelas contíguas e o alvo s[i] (em escala por série)."""
    X, y, meta = [], [], []
    for (p, pr), g in df_src.groupby(['pdv','produto'], sort=False):
        g = g.sort_values('week_start')
        s    = g['quantity_scaled'].to_numpy(np.float32)
        ymin = g['ymin'].to_numpy(np.float32)
        yrng = g['range'].to_numpy(np.float32)
        n = len(s)
        if n <= lookback:
            continue
        for i in range(lookback, n):
            X.append(s[i-lookback:i].reshape(lookback,1))
            y.append(s[i])
            meta.append((float(ymin[i]), float(yrng[i])))
    if not X:
        return None, None, None
    return np.stack(X), np.array(y, np.float32).reshape(-1,1), meta


# Muito parecida com a anterior, mas ela só cria exemplos onde o próximo ponto (o alvo) está no período de teste.
def build_windows_cross_boundary_with_meta(df_all, lookback=8):
    """Teste cross-boundary: entradas podem vir do treino, alvo é a primeira semana em 'test' e seguintes."""
    X, y, meta = [], [], []
    for (p, pr), g in df_all.groupby(['pdv','produto'], sort=False):
        g = g.sort_values('week_start')
        s     = g['quantity_scaled'].to_numpy(np.float32)
        splits= g['split'].to_numpy()
        ymin  = g['ymin'].to_numpy(np.float32)
        yrng  = g['range'].to_numpy(np.float32)
        n = len(s)
        for i in range(lookback, n):
            if splits[i] != 'test':
                continue
            win = s[i-lookback:i]
            if np.any(np.isnan(win)):
                continue
            X.append(win.reshape(lookback,1))
            y.append(s[i])
            meta.append((float(ymin[i]), float(yrng[i])))
    if not X:
        return None, None, None
    return np.stack(X), np.array(y, np.float32).reshape(-1,1), meta



# Pega a meta_list e organiza em dois vetores/colunas (ymin e range) para ficar fácil de fazer conta depois.
def meta_to_arrays(meta_list):
    ymin = np.array([m[0] for m in meta_list], dtype=np.float64).reshape(-1,1)
    yrng = np.array([m[1] for m in meta_list], dtype=np.float64).reshape(-1,1)
    return ymin, yrng


# desfaz a escala para voltar ao valor real
def inverse_per_series_from_meta(y_scaled, meta_list):
    ymin, yrng = meta_to_arrays(meta_list)
    return y_scaled.astype(np.float64) * yrng + ymin


# calcula o wmape para avaliar o modelo
def wmape(y_true, y_pred):
    y_true = np.asarray(y_true).reshape(-1)
    y_pred = np.asarray(y_pred).reshape(-1)
    denom = np.abs(y_true).sum()
    return np.nan if denom == 0 else np.abs(y_true - y_pred).sum() / denom


def clean_windows(X, y, meta):
    """Remove qualquer janela com NaN/Inf em X, y ou meta."""
    if X is None or y is None or meta is None:
        return None, None, None
    ymin, yrng = meta_to_arrays(meta)
    m = np.isfinite(X).all(axis=(1,2)) & np.isfinite(y).ravel() \
        & np.isfinite(ymin).ravel() & np.isfinite(yrng).ravel()
    if not m.any():
        return None, None, None
    Xc = X[m]
    yc = y[m]
    meta_c = [meta[i] for i, ok in enumerate(m) if ok]
    return Xc, yc, meta_c


df_nn_safe = df_nn.copy()
if 'range' in df_nn_safe.columns:
    df_nn_safe['range'] = df_nn_safe['range'].replace(0, 1.0)

MIN_WEEKS_PER_PAIR = 16
weeks_per_pair = df_nn_safe.groupby(['pdv','produto']).size().rename('n').reset_index()
keep_pairs = weeks_per_pair[weeks_per_pair['n'] >= MIN_WEEKS_PER_PAIR][['pdv','produto']]
df_nn_safe = df_nn_safe.merge(keep_pairs, on=['pdv','produto'], how='inner')

# Criando dataframes apenas com as colunas que iremos usar para rodar mais rápido
df_tr_small  = df_nn_safe[df_nn_safe['split']=='train'][['pdv','produto','week_start','quantity_scaled','ymin','range']]
df_all_small = df_nn_safe[['pdv','produto','week_start','quantity_scaled','split','ymin','range']]

print(f"[STAGE 4] Preparando janelas…", flush=True)
```

### Treinando efetivamente nosso modelo

```python
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, callbacks, losses
tf.keras.utils.set_random_seed(RANDOM_SEED)

def build_model(lookback, units, dropout, lr, loss_name):
    m = models.Sequential([
        layers.Input(shape=(lookback, 1)),
        layers.LSTM(units, dropout=dropout),
        layers.Dense(units//2, activation='relu'),
        layers.Dense(1, activation='linear')
    ])
    loss_fn = losses.Huber(delta=1.0) if loss_name=='huber' else losses.MeanAbsoluteError()
    m.compile(optimizer=optimizers.Adam(lr), loss=loss_fn, metrics=['mae'])
    return m

MAX_TRAIN_WINDOWS = 800_000   # Ponha None para remover o cap
CAP_SEED = 42

best = {'wmape': np.inf, 'cfg': None, 'baseline': None}
best_model = None
best_lookback = None

for cfg in CONFIGS:
    lookback   = cfg['lookback']
    units      = cfg['units']
    dropout    = cfg['dropout']
    loss_name  = cfg['loss']
    lr         = cfg['lr']
    batch_size = cfg['batch_size']
    epochs     = cfg['epochs']

    print(f"\n[TRY] lb={lookback} | units={units} | drop={dropout} | loss={loss_name} | lr={lr} | batch={batch_size} | ep={epochs}", flush=True)

    # Diagnóstico do nº de janelas de treino (estimativa)
    grp_sizes = (df_tr_small.groupby(['pdv','produto'], sort=False).size().to_numpy())
    est_tr_windows = np.maximum(grp_sizes - lookback, 0).sum()
    pairs_tr = df_tr_small[['pdv','produto']].drop_duplicates().shape[0]
    pairs_te = df_all_small[['pdv','produto']].drop_duplicates().shape[0]
    print(f"[WIN] pares treino: {pairs_tr:,} | pares total (p/ teste): {pairs_te:,}")
    print(f"[WIN] estimativa de janelas treino: {est_tr_windows:,}")

    # Monta janelas
    tJ = time.time()
    X_tr, y_tr, meta_tr = build_windows_intrasplit_with_meta(df_tr_small, lookback=lookback)
    X_te, y_te, meta_te = build_windows_cross_boundary_with_meta(df_all_small, lookback=lookback)

    # Limpeza anti-NaN/Inf
    X_tr, y_tr, meta_tr = clean_windows(X_tr, y_tr, meta_tr)
    X_te, y_te, meta_te = clean_windows(X_te, y_te, meta_te)
    if X_tr is None or X_te is None:
        print(" -> janelas insuficientes/invalidas após limpeza; pulando.", flush=True)
        continue

    # Cap determinístico 
    if MAX_TRAIN_WINDOWS is not None and len(X_tr) > MAX_TRAIN_WINDOWS:
        rng = np.random.RandomState(CAP_SEED)
        idx = rng.choice(len(X_tr), size=MAX_TRAIN_WINDOWS, replace=False)
        X_tr, y_tr = X_tr[idx], y_tr[idx]
        meta_tr = [meta_tr[i] for i in idx]

    print(f"   janelas: X_tr={X_tr.shape} | X_te={X_te.shape} | {time.time()-tJ:.1f}s", flush=True)

    # Pesos ~ |y_real| (proxima WMAPE) com proteção
    ymin_tr, yrng_tr = meta_to_arrays(meta_tr)
    y_tr_real = y_tr.astype(np.float64) * yrng_tr + ymin_tr
    sample_weight = np.abs(y_tr_real.ravel())
    sample_weight = np.nan_to_num(sample_weight, nan=0.0, posinf=0.0, neginf=0.0)
    if sample_weight.sum() == 0:
        sample_weight = np.ones_like(sample_weight, dtype=np.float64)
    else:
        sample_weight = sample_weight / (sample_weight.mean() + 1e-9)

    # Modelo + callbacks
    model = build_model(lookback, units, dropout, lr, loss_name)
    es    = callbacks.EarlyStopping(monitor='val_mae', patience=2, restore_best_weights=True)
    rlrop = callbacks.ReduceLROnPlateau(monitor='val_mae', factor=0.5, patience=1, min_lr=1e-5)

    tT = time.time()
    history = model.fit(
        X_tr, y_tr,
        validation_data=(X_te, y_te),
        sample_weight=sample_weight,
        epochs=epochs,
        batch_size=batch_size,
        verbose=1,
        callbacks=[es, rlrop]
    )
    print(f"   treino: {time.time()-tT:.1f}s", flush=True)

    # Avaliação em unidade real
    y_te_hat_scaled = model.predict(X_te, verbose=0)
    y_te_real = inverse_per_series_from_meta(y_te, meta_te)
    y_te_hat  = inverse_per_series_from_meta(y_te_hat_scaled, meta_te)

    m = wmape(y_te_real, y_te_hat)

    # baseline lag-1
    y_hat_lag1_scaled = X_te[:, -1, :]
    y_hat_lag1 = inverse_per_series_from_meta(y_hat_lag1_scaled, meta_te)
    m_base = wmape(y_te_real, y_hat_lag1)

    m_txt = "N/A" if not np.isfinite(m) else f"{m:.4%}"
    b_txt = "N/A" if not np.isfinite(m_base) else f"{m_base:.4%}"
    print(f"   -> WMAPE(test)={m_txt} | baseline={b_txt}", flush=True)

    # Seleciona o melhor
    if np.isfinite(m) and (m < best['wmape']):
        best.update({'wmape': m,
                     'cfg': {**cfg, 'epochs_trained': len(history.history['loss'])},
                     'baseline': m_base if np.isfinite(m_base) else None})
        if best_model is not None:
            try: del best_model
            except: pass
            gc.collect()
        best_model   = model
        best_lookback= lookback
    else:
        del model
        gc.collect()

    # limpa temporários pesados
    del X_tr, y_tr, meta_tr, X_te, y_te, meta_te, y_te_hat_scaled, y_te_hat, y_te_real, y_hat_lag1_scaled, y_hat_lag1
    gc.collect()

print("\n==================== RESULTADO FINAL (FAST MODE) ====================")
print("Melhor WMAPE(test):", "N/A" if not np.isfinite(best['wmape']) else f"{best['wmape']:.4%}")
base_txt = "N/A" if (best['baseline'] is None or not np.isfinite(best['baseline'])) else f"{best['baseline']:.4%}"
print("Baseline lag-1    :", base_txt)
print("Melhor config     :", best['cfg'])
```

### Vamos agora gerar o csv com as previsões

```python
CSV_PATH = 'previsoes_jan2023.csv'
H = 5  # 5 semanas de janeiro (ISO 1..5)

# última semana observada no treino
last_train_week = df_nn_safe.loc[df_nn_safe['split']=='train','week_start'].max()

# dataframe mínimo + flag histórico (<= fim do treino)
meta_cols = ['pdv','produto','week_start','quantity_scaled','ymin','range']
df_meta = df_nn_safe[meta_cols].copy()
df_meta['is_hist'] = (df_meta['week_start'] <= last_train_week)

# construindo janela inicial por (pdv, produto)
init_windows, ymins, yrngs, pairs = [], [], [], []
for (p, pr), g in df_meta.groupby(['pdv','produto'], sort=False):
    gh = g[g['is_hist']].sort_values('week_start')
    if gh.empty:
        continue
    s = gh['quantity_scaled'].to_numpy(np.float32)
    # padding se histórico < lookback (repete último valor)
    if s.size >= best_lookback:
        w = s[-best_lookback:]
    else:
        w = np.concatenate([np.full(best_lookback - s.size, s[-1], dtype=np.float32), s])
    init_windows.append(w.reshape(best_lookback, 1))
    # min/max (fit no treino) da série
    ymins.append(float(gh['ymin'].iloc[-1]))
    yrngs.append(float(gh['range'].iloc[-1]))
    pairs.append((p, pr))

if not init_windows:
    raise RuntimeError("Nenhum par com histórico suficiente para prever.")

W = np.stack(init_windows, axis=0) 
ymin_arr = np.array(ymins, dtype=np.float64).reshape(-1,1)
yrng_arr = np.array(yrngs, dtype=np.float64).reshape(-1,1)
pairs_used = pd.DataFrame(pairs, columns=['pdv','produto'])

print(f"[FORECAST] Séries: {W.shape[0]:,} | Lookback: {best_lookback} | Rollout: {H} semanas")

# rollout autoregressivo em lote (5 passos)
BATCH_PRED = 4096
preds_scaled_steps = []
for step in range(H):
    yhat = best_model.predict(W, batch_size=BATCH_PRED, verbose=0)  
    preds_scaled_steps.append(yhat.astype(np.float32))
    W = np.concatenate([W[:, 1:, :], yhat.reshape(-1,1,1)], axis=1)
    print(f"  passo {step+1}/{H} concluído")

# desescalando e montando DataFrame final
preds_scaled = np.hstack(preds_scaled_steps)                
preds_real   = preds_scaled.astype(np.float64) * yrng_arr + ymin_arr  

out_chunks = []
for k in range(H):  # semanas 1..H
    tmp = pairs_used.copy()
    tmp['semana'] = k + 1
    tmp['quantidade'] = preds_real[:, k]
    out_chunks.append(tmp[['semana','pdv','produto','quantidade']])

df_out = pd.concat(out_chunks, ignore_index=True)


df_out.to_csv(CSV_PATH, index=False, sep=';', encoding='utf-8')
print(f"[CSV] Arquivo gerado com {len(df_out):,} linhas -> {CSV_PATH}")
print(df_out.head())
```
